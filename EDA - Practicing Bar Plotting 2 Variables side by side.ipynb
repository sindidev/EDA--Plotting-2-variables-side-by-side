```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.DataFrame({
    "Year": ["2004", "2005", "2006", "2007"],
    "Revenue for camp gear": [17, 38, 66, 43],
    "Predicted revenue for camp gear": [14, 39, 64, 45]})
df.plot.bar() 
plt.bar(df['Predicted revenue for camp gear'], df['Revenue for camp gear']) 
plt.xlabel("Year") 
plt.ylabel("Revenue in Millions") 
plt.show() 
```
